# Data Preparation

### Importing the importance libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Loading .csv dataset for data preparation

```python
### specify the absolute path of dataset
dataset_path = "../dataset/ads.csv"

### importing the dataset
df = pd.read_csv(dataset_path)
```

### Data Preview

```python
df.sample(3, random_state=42)
```

## 1. Handling missing values

```python
df.isna().sum()
```

##### As we can see there is no any missing values in this dataset.

## 2. Outlier Detection

### Outlier visualization of TV

```python
### Plot the distribution boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df["TV"])
sns.stripplot(data=df["TV"])
plt.show()
```

##### As we can see data are between lower and upper fence

## Outlier Detection of radio

```python
### Plot the distribution boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df["radio"])
sns.stripplot(data=df["radio"])
plt.show()
```

##### As we can see data are between lower and upper fence

## Outlier Detection of newspaper

```python
### Plot the distribution boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df["newspaper"])
sns.stripplot(data=df["newspaper"])
plt.show()
```

##### As we can see most of the data are between lower and upper fence but few seem above upper fence

## Outlier Detection of sales

```python
### Plot the distribution boxplot
plt.figure(figsize=(8,6))
sns.boxplot(data=df["sales"])
sns.stripplot(data=df["sales"])
plt.show()
```

##### As we can see data are between lower and upper fence

# With no missing values and no outliers detected in any of the columns, the dataset is now ready for model preparation and implementation.

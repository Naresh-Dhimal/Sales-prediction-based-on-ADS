# Exploratory Data Analysis 

### Importing importance libraries

```python
### for data analysis
import pandas as pd

### for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing dataset

```python
### Specify the absoulate path
dataset_path = "../dataset/ads.csv"

### loading .csv dataset and converting to pandas DataFrame
df = pd.read_csv(dataset_path)
```

### Data Preview

```python
df.head(3)
```

### 1. Univariate Analysis

```python
#checking the columns present in dataset
df.columns
```

# TV

```python
# to check the distribution of Pregnancies (for numerical features we use .histplot() or .displot())

# Plot the distribution
plt.figure(figsize=(10,10))
sns.displot(data=df["TV"],kde=True)
plt.title("Distribution plot of TV")
plt.show()
```

### Based on the distribution plot above for the variable "TV":
* The majority of data points fall within the range of 200 to 230.

# radio

```python
# plot the distribution 
plt.figure(figsize=(10,10))
sns.displot(data=df["radio"], kde=True)
plt.title("Distribution plot of radio")
plt.show()
```

### Based on the distribution plot above for the variable "radio":
* The distribution plot indicates that the data is approximately normally distributed.
* The majority of data points fall within the range of 0 to 5 with over 30 counts and within the range 35 to 45 with over 25.

# newspaper

```python
# plot the distribution 
plt.figure(figsize=(10,10))
sns.displot(data=df["newspaper"], kde=True)
plt.title("Distribution plot of newspaper")
plt.show()
```

### Based on the distribution plot above for the variable "newspaper":
* It can be concluded that the data is right-skewed.
* The majority of data points fall within the range of 0 to 80.

# sales

```python
# plot the distribution 
plt.figure(figsize=(10,10))
sns.displot(data=df["sales"], kde=True)
plt.title("Distribution plot of sales")
plt.show()
```

### Based on the distribution plot above for the variable "sales":
* The distribution plot indicates that the data is approximately normally distributed.
* The majority of data points fall within the range of 5 to 25.

# 2. Bivariate Analysis

# TV vs sales

```python
# to check the distribution of numerical vs numerical feature .scatterplot() is used

# plot the distribution
plt.figure(figsize=(10,6))
sns.scatterplot(x="TV", y="sales", data=df, color="purple")
plt.title("Distribution of TV vs sales")
plt.show()
```

### From above observation,
* There appears to be a positive correlation between the investment in TV advertisements and sales. As the expenditure on TV ads increases, there is a corresponding upward trend in sales. This suggests that TV advertising may have a significant impact on driving sales figures higher. 


# radio vs sales

```python
# plot the distribution
plt.figure(figsize=(10,6))
sns.scatterplot(x="radio", y="sales", data=df, color="purple")
plt.title("Distribution of radio vs sales")
plt.show()
```

### From above observation,
* Sometimes, an increase in radio advertising expenditure coincides with a decrease in sales, while in other instances, a decrease in radio ads aligns with a decrease in sales as well. This suggests that the impact of radio advertising on sales may not be as straightforward and consistent.

# newspaper vs sales

```python
# plot the distribution
plt.figure(figsize=(10,6))
sns.scatterplot(x="newspaper", y="sales", data=df, color="purple")
plt.title("Distribution of newspaper vs sales")
plt.show()
```

### From above observation,
* The relationship between newspaper advertising and sales data seems to exhibit a more scattered pattern. There isn't a clear trend indicating a direct correlation between newspaper ad spending and sales figures.
